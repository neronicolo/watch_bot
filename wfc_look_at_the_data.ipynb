```python
import check
from pathlib import Path
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
```

```python
dir_path = check.path_check(Path.home()/'programming/data/watch_bot/')
df = pd.read_csv(dir_path/'wfc_file_attribs.csv')
#df.head()
print(f'{df.shape}\n{df.dtypes}')
```

```python
# Remove rows where image_size_x and image_size_y are less than size.
size = 300
df = df[(df['image_size_x'] > size) & (df['image_size_y'] > size)]
df.shape
```

```python
# Plot image_size_x next to image_size_y. On one canvas.
fig, axes = plt.subplots(1, 2, figsize=(10,4), tight_layout=True)
axes[0].hist(df['image_size_x']);
axes[1].hist(df['image_size_y']);
```
